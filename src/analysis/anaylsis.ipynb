Análise de duplicatas

```python
# Total de linhas e quantidade de valores distintos por coluna
print(f"Total de linhas: {len(df)}")
df.nunique()
```

```python
# Para cada par de colunas (A, B): das linhas duplicadas em A, qual % também é duplicada em B
dup_masks = {col: df.duplicated(subset=[col], keep=False) for col in df.columns}

pd.DataFrame({
    col_b: {
        col_a: (mask_a & dup_masks[col_b]).sum() / mask_a.sum() if mask_a.sum() else float("nan")
        for col_a, mask_a in dup_masks.items()
    }
    for col_b in df.columns
})
```

Consistência de category nos registros duplicados (title/text)

```python
dup_text_categories = df[df.duplicated(subset=["text"], keep=False)]\
                        .groupby("text")["category"]\
                        .nunique()
                        
print(f"Total: {len(dup_text_categories)}")
print(f"Categoria diferentes: {(dup_text_categories > 1).sum()}")
```

```python
dup_title_categories = df[df.duplicated(subset=["title"], keep=False)]\
                       .groupby("title")\
                       ["category"].nunique()

print(f"Total: {len(dup_title_categories)}")
print(f"Categoria diferentes: {(dup_title_categories > 1).sum()}")
```

```python
dup_title_text_categories = df[df.duplicated(subset=["title", "text"], keep=False)]\
                            .groupby(["title", "text"])["category"]\
                            .nunique()

print(f"Total: {len(dup_title_text_categories)}")
print(f"Categoria diferentes: {(dup_title_text_categories > 1).sum()}")
```

Verificando poluição no texto (tags HTML, entidades, espaços, URLs)

```python
padroes_poluicao = {
    "tags_html": r"<[^>]+>",
    "entidades_html": r"&[a-zA-Z]+;|&#\d+;",
    "espacos_multiplos": r"\s{2,}",
    "urls": r"https?://\S+",
}

pd.Series({
    nome: df["text"]\
            .fillna("")\
            .str.contains(regex, regex=True)\
            .sum()
            
    for nome, regex in padroes_poluicao.items()
})
```

```python
# 5 exemplos de texto pra cada tipo de poluição encontrado
with pd.option_context("display.width", None, "display.max_colwidth", None):
    for nome, regex in padroes_poluicao.items():
        print(f"--- {nome} ---")
        display(df[df["text"]\
                .str.contains(regex, regex=True)][["title", "text"]]\
                .head(5)
        )
```
